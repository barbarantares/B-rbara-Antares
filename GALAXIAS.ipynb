```python
import fsps
import matplotlib.pyplot as plt
import numpy as np
```

```python
sp = fsps.StellarPopulation(imf_type=3,zcontinuous=1, imf_upper_limit=100, imf_lower_limit=0.1, sfh=1,tau=3,sf_start=0,sf_trunc=0, fburst = 0, tburst = 11, const = 0.3,dust_type=3, dust1=1.0, dust2=0.5)
sp.params['logzsol'] = 0.2
sp.params['gas_logz'] = 0.2 #Metalicidad
sp.params['gas_logu'] = -3
wave, spec = sp.get_spectrum(tage=10,peraa=True)
plt.figure(figsize=(10, 6))
plt.plot(wave, spec, label='Espectro con líneas de emisión')
plt.xlabel("Wavelength (Å)")
plt.ylabel("log($F_\lambda \, (L_{\odot} \, \mathrm{\AA}^{-1} \, M_{\odot}^{-1})$)")
plt.title('Spiral Galaxy Spectrum')
plt.xlim(3000, 10000)
plt.legend()
plt.savefig('espiral.jpg', format='jpg',dpi=300)
plt.show()
```
